```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bos1 = pd.read_csv("BostonHousing.csv")
bos1
```

```python
bos1.columns
```

```python
x = bos1[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax','ptratio', 'b', 'lstat']]
y = bos1["medv"]
```

```python
sns.heatmap(bos1.corr())
plt.show()
```

```python
bos1.corr()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression #step 1 : Import
lm = LinearRegression() #step 2: Instantiate: Create an object 
lm.fit(X_train,y_train) #step 3 : Fit/Learn
pred_y = lm.predict(X_test) #step 4: Predict
```

```python
plt.scatter(y_test,pred_y)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test,pred_y)
mse
```

# Second Model

```python
x = bos1[['crim', 'zn', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'b']]
y = bos1["medv"]
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.20)

from sklearn.linear_model import LinearRegression #step 1 : Import
lm = LinearRegression() #step 2: Instantiate: Create an object 
lm.fit(X_train,y_train) #step 3 : Fit/Learn
pred_y = lm.predict(X_test) #step 4: Predict

mse = mean_squared_error(y_test,pred_y)
mse
```
