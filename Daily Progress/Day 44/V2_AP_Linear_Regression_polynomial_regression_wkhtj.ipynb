```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bos1 = pd.read_csv("BostonHousing.csv")
bos1
```

```python
bos1.columns
```

```python
x = bos1[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax','ptratio', 'b', 'lstat']]
y = bos1["medv"]
```

```python
sns.heatmap(bos1.corr())
plt.show()
```

```python
bos1.corr()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression #step 1 : Import
lm = LinearRegression() #step 2: Instantiate: Create an object 
lm.fit(X_train,y_train) #step 3 : Fit/Learn
pred_y = lm.predict(X_test) #step 4: Predict
```

```python
plt.scatter(y_test,pred_y)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test,pred_y)
mse
```

# Second Model

```python
x = bos1[['crim', 'zn', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'b']]
y = bos1["medv"]
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.20)

from sklearn.linear_model import LinearRegression #step 1 : Import
lm = LinearRegression() #step 2: Instantiate: Create an object 
lm.fit(X_train,y_train) #step 3 : Fit/Learn
pred_y = lm.predict(X_test) #step 4: Predict

mse = mean_squared_error(y_test,pred_y)
mse
```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures #step 1: Import
```

```python
poly = PolynomialFeatures(degree=2) #step 2
poly_features = poly.fit_transform(x)
```

```python
poly_features
```

```python
X_train,X_test,y_train,y_test = train_test_split(poly_features,y,test_size=0.20)
```

```python
poly_reg = LinearRegression()
```

```python
poly_reg.fit(X_train,y_train)
```

```python
poly_reg.coef_
```

```python
y_pred = poly_reg.predict(X_test)
y_pred
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

# Decision Trees

```python
x = bos1[['crim', 'zn', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'b']]
y = bos1["medv"]
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.20)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regdt = DecisionTreeRegressor()
```

```python
regdt.fit(X_train,y_train)
```

```python
y_pred = regdt.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
r2_score(y_test,y_pred)
```

# Random Forest Regressor


```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regrf = RandomForestRegressor(n_estimators=500)
```

```python
regrf.fit(X_train,y_train)
```

```python
y_predrf = regrf.predict(X_test)
y_predrf
```

```python
np.sqrt(mean_squared_error(y_test,y_predrf))
```

```python
r2_score(y_test,y_predrf)
```
